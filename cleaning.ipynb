### 2022

```python
# make all the imports 

import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
import re
from playwright.async_api import async_playwright
```

```python
# read csv

df_2022 = pd.read_csv('mass-shootings-2022.csv')
df_2022
```

```python
# rename columna

df_2022.rename(columns={'Incident ID': 'incident_id', 'Incident Date': 'incident_date', 'City Or County': 'city_or_country', '# Killed': 'killed', '# Injured': 'injured', 'Address': 'address', 'State': 'state'}, inplace=True)
```

```python
# remove column Operations 

# df_2022.drop(columns=['Operations'], inplace=True)
# df_2022
```

```python
# create a new column 'incident_url' which is https://www.gunviolencearchive.org/ + incident_id

df_2022['incident_url'] = 'https://www.gunviolencearchive.org/incident/' + df_2022['incident_id'].astype(str)
df_2022
```

```python
df_2022.dtypes
```

```python
# convert incident_date to datetime

df_2022['incident_date'] = pd.to_datetime(df_2022['incident_date'])
df_2022.dtypes
```

```python
# create a new column 'year' which is the year of the incident

df_2022['year'] = df_2022['incident_date'].dt.year
df_2022
```

```python
# create a new column 'month-year' column 

df_2022['month-year'] = df_2022['incident_date'].dt.strftime('%m-%Y')
df_2022
```

```python
# remove columns index and level_0

df_2022.drop(columns=['index', 'level_0'], inplace=True)
```

```python
# reset index 

df_2022.reset_index(inplace=True)
df_2022
```

```python
# save csv

df_2022.to_csv('mass-shootings-2022.csv', index=False)
```

```python
shootings_2013 = []

while True:
    try: 
        html = await page.content()
        soup = BeautifulSoup(html, 'html.parser')
        data = []
        for tr in soup.find_all('tr'):
            row = {}
            tds = tr.find_all('td')
            if len(tds) > 0:
                row["incident_id"] = tds[0].text
                row["incident_date"] = tds[1].text
                row["state"] = tds[2].text
                row["city_or_county"] = tds[3].text
                row["address"] = tds[4].text
                row["killed"] = tds[5].text
                row["injured"] = tds[6].text
                row["operations"] = tds[7].text
                shootings_2013.append(row)

        async with page.expect_navigation():
            # locate the next button
            await page.locator('//*[@id="block-system-main"]/div/ul/li[11]/a').click()
        print("Pressing the button")
        print("---------")
    except: # Exception as e:
        # print the error and break out of the loop
        #print(f"An error occurred: {e}")
        break

shootings_2013
```
