```python
# make all the imports 

import pandas as pd
```

```python
# read csv

df = pd.read_csv('mass-shootings-2022.csv')
df
```

```python
# rename columna

df.rename(columns={'Incident ID': 'incident_id', 'Incident Date': 'incident_date', 'City Or County': 'city_or_country', '# Killed': 'killed', '# Injured': 'injured', 'Address': 'address', 'State': 'state'}, inplace=True)
```

```python
# remove column Operations 

df.drop(columns=['Operations'], inplace=True)
```

```python
df
```

```python
# create a new column 'incident_url' which is https://www.gunviolencearchive.org/ + incident_id

df['incident_url'] = 'https://www.gunviolencearchive.org/incident/' + df['incident_id'].astype(str)
df
```

```python
df.dtypes
```

```python
# convert incident_date to datetime

df['incident_date'] = pd.to_datetime(df['incident_date'])
df.dtypes
```

```python
# create a new column 'year' which is the year of the incident

df['year'] = df['incident_date'].dt.year
df
```

```python
# create a new column 'month-year' colum 

df['month-year'] = df['incident_date'].dt.strftime('%m-%Y')
df
```
